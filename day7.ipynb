```python
dir_structure = {}
current_path = []
        
with open('day7.txt', 'r', encoding='utf-8') as file:
    for line in file:
        if line.startswith("$ cd"):
            _, _, path = line.split()
            if path == "/":
                current_path = []
            elif path == "..":
                if current_path:
                    current_path.pop()
            else:
                current_path.append(path)
        elif line.startswith("$ ls"):
            continue
        else:
            size_or_dir, name = line.rsplit(maxsplit=1)
            current_dir = dir_structure
            for dir in current_path:
                current_dir = current_dir.setdefault(dir, {})
            if size_or_dir == "dir":
                current_dir[name] = {}
            else:
                current_dir[name] = int(size_or_dir)
                
dir_structure
```

```python
def calculate_total_size(directory, path=""):
    total_size = 0
    sum_sizes = 0
    sizes = []

    for key, value in directory.items():
        if isinstance(value, dict):
            dir_size, dir_sum, dir_sizes = calculate_total_size(value, path + "/" + key)
            # print(f"Total size of directory '{path + '/' + key}': {dir_size}")
            total_size += dir_size
            sum_sizes += dir_sum
            sizes.extend(dir_sizes)
        else:
            total_size += value

    if total_size <= 100000:
        sum_sizes += total_size
    sizes.append(total_size)
    return total_size, sum_sizes, sizes
```

```python
total, sum_below_limit, all_sizes = calculate_total_size(dir_structure)
print(f"Total size below limit: {sum_below_limit}")
x = 70000000 - total
y = 30000000 - x
print("Free disk:",x)
print("Need more:",y)
```

```python
min_size = 9999999999
for size in all_sizes:
    if y < size < min_size:
        min_size = size
        
min_size
```
